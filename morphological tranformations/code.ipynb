```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# erosion

```python
img=cv2.imread("./j.png",0)
kernel=np.ones((5,5),np.uint8)
erosion=cv2.erode(img,kernel,iterations=1)
plt.subplot(121),plt.imshow(img,'gray');plt.title("orignal image")
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(erosion, 'gray');plt.title("erorde image")
plt.xticks([]),plt.yticks([])
```

# dilation

```python
# img=cv2.imread("./j.png",0)
# kernel=np.ones((5,5),np.uint8)
dilation=cv2.dilate(erosion,kernel,iterations=1)
plt.subplot(121),plt.imshow(img,'gray');plt.title("orignal image")
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(dilation, 'gray');plt.title("dilation")
plt.xticks([]),plt.yticks([])
```

```python
img=cv2.imread("./j.png",0)
img=img/255.0
random=np.random.randint(0,2,(img.shape[0],img.shape[1]))

random=random.astype(np.float64)

bit=cv2.bitwise_or(img,random)

# plt.imshow(bit,'gray')
```

```python
random=np.random.random((img.shape[0],img.shape[1]))

ret,th1=cv2.threshold(random,0.99,1,cv2.THRESH_BINARY)
bit=cv2.bitwise_or(img,th1)
# bit=img+th1
plt.imshow(bit,'gray')
```

```python
opening=cv2.morphologyEx(bit,cv2.MORPH_OPEN,kernel)
plt.subplot(121),plt.imshow(bit,'gray');plt.title("orignal image")
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(opening, 'gray');plt.title("opening")
plt.xticks([]),plt.yticks([])
```
