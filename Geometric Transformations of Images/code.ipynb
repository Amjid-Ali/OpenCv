# translation

```python
img=cv2.imread("../messie.jpeg",0)
rows,cols = img.shape
M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# rotations

```python
img=cv2.imread("../messie.jpeg",0)
rows,cols=img.shape


# the arugments to the function is center, angle and scale(zoom)
rotate=cv2.getRotationMatrix2D(((cols-1)/2.0,(rows-1)/2.0),180,4)
dst=cv2.warpAffine(img,rotate, (cols,rows))
cv2.imshow("img",dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# affine transfomration

```python
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("../messie.jpeg",0)
rows,cols=img.shape


pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])

M = cv2.getAffineTransform(pts1,pts2)
```

```python
dst = cv2.warpAffine(img,M,(cols,rows))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

# Perspective Transformation

```python
img=cv2.imread("../messie.jpeg",0)
rows,cols=img.shape

pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])


M = cv2.getPerspectiveTransform(pts1,pts2)

dst=cv2.warpPerspective(img,M,(300,300))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
