```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("./blox.jpg")
gray_scale=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
corners=cv2.goodFeaturesToTrack(gray_scale,50,0.01,10)
corners=np.uint0(corners)
```

```python
for x in corners:
    i,j=x.ravel()
    cv2.circle(image,(i,j),1,255,1)
plt.imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
