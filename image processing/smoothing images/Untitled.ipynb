```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# custom filters

```python
path=r"D:\Amjid_Ali\opencv\OpenCv\core operations on image\basic operations on images\opencv-logo.png"

img=cv2.imread(path)
kernel=np.ones((5,5),np.float32)/25
kernel_5=np.ones((7,7),np.float32)/35
kernel_11=np.ones((11,11),np.float32)/44

dst=cv2.filter2D(img,-1,kernel)
dst5=cv2.filter2D(img,-1,kernel_5)
dst11=cv2.filter2D(img,-1,kernel_11)
plt.subplot(121),plt.imshow(img),plt.title("original")
# plt.subplot(122),plt.imshow(dst),plt.title("Average")
# plt.subplot(122),plt.imshow(dst5),plt.title("Average 7x7 kernel")
plt.subplot(122),plt.imshow(dst11),plt.title("Average 11x11 kernel")
plt.show()
```

# built-in function 

```python
# average ->blur
```

```python
img=cv2.imread(path)
blur=cv2.blur(img,(13,13))
plt.subplot(121),plt.imshow(img),plt.title("original")
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title("Average")
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
# Gaussian Blur 
```

```python
img=cv2.imread(path)
blur=cv2.GaussianBlur(img,(13,13),0)
plt.subplot(121),plt.imshow(img),plt.title("original")
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title("Average")
plt.xticks([]),plt.yticks([])
plt.show()
```
