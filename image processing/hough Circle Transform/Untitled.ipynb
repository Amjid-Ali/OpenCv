```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("./opencv.png",0)
```

```python
plt.imshow(image,'gray')
```

```python
# blur it first
blur_image=cv2.medianBlur(image,5)
```

```python
color_image=cv2.cvtColor(blur_image,cv2.COLOR_GRAY2BGR)
circles = cv2.HoughCircles(blur_image,cv2.HOUGH_GRADIENT,1,20,
                            param1=50,param2=30,minRadius=0,maxRadius=0)
```

```python
circles
```

```python
circle=np.uint16(np.around(circles))
```

```python
for i in circle[0,:]:
    cv2.circle(color_image,(i[0],i[1]),i[2],(255,0,255),2)
    cv2.circle(color_image,(i[0],i[1]),1,(0,0,255),2)
cv2.imshow("image",color_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
